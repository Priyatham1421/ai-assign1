````python
def is_armstrong(number):
  """Checks if a number is an Armstrong number.

  An Armstrong number is a number that is equal to the sum of its own digits each
  raised to the power of the number of digits.

  Args:
    number: An integer to check.

  Returns:
    True if the number is an Armstrong number, False otherwise.
  """
  # Convert the number to a string to get the number of digits
  num_str = str(number)
  num_digits = len(num_str)

  sum_of_powers = 0
  # Iterate through each digit
  for digit_char in num_str:
    # Convert the digit back to an integer
    digit = int(digit_char)
    # Calculate the digit raised to the power of the number of digits
    sum_of_powers += digit ** num_digits

  # Compare the original number with the sum of powers
  return number == sum_of_powers

# Example usage
test_number = 153
if is_armstrong(test_number):
  print(f"{test_number} is an Armstrong number.")
else:
  print(f"{test_number} is not an Armstrong number.")

test_number = 123
if is_armstrong(test_number):
  print(f"{test_number} is an Armstrong number.")
else:
  print(f"{test_number} is not an Armstrong number.")
````

````python
def is_armstrong_optimized(number):
  """Checks if a number is an Armstrong number without string conversion.

  An Armstrong number is a number that is equal to the sum of its own digits each
  raised to the power of the number of digits.

  Args:
    number: An integer to check.

  Returns:
    True if the number is an Armstrong number, False otherwise.
  """
  original_number = number
  num_digits = 0
  temp_number = number

  # Count the number of digits
  while temp_number > 0:
    temp_number //= 10
    num_digits += 1

  sum_of_powers = 0
  temp_number = number # Reset temp_number for digit extraction

  # Calculate the sum of powers of digits
  while temp_number > 0:
    digit = temp_number % 10
    sum_of_powers += digit ** num_digits
    temp_number //= 10

  return original_number == sum_of_powers

# Example usage
test_number = 153
if is_armstrong_optimized(test_number):
  print(f"{test_number} is an Armstrong number.")
else:
  print(f"{test_number} is not an Armstrong number.")

test_number = 123
if is_armstrong_optimized(test_number):
  print(f"{test_number} is an Armstrong number.")
else:
  print(f"{test_number} is not an Armstrong number.")

test_number = 9474
if is_armstrong_optimized(test_number):
  print(f"{test_number} is an Armstrong number.")
else:
  print(f"{test_number} is not an Armstrong number.")
````
